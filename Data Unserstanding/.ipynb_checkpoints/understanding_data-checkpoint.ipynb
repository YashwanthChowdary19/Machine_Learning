```python
df = pd.read_csv("tmdb_movies_2021_2025.csv")
```

##  <b style="color:yellow"> Seven Important Questions to understand dataset

### How big the data is ? 

```python
# to get the number of rows and columns pandas provides attribute called shape 
df.shape
```

### How does the data looks like ?

```python
df.sample(5) # always use sample() instead of head() or tail() bcz it gives the random rows it is best understand the data
```

### What is the datatypes of columns ? 

```python
df.info() # returns the number of number of not null values and datatypes of each columns
```

### Are there any missing values ?

```python
# to get missing values 
df.isnull().sum() # returns the number of null values for each column
```

### How does the data looks like mathematically ? 

```python
df.describe() # returns the mathematical concepts like below for only numerical columns
```

### Are there any duplicates values ?

```python
df.duplicated().sum() # there is no duplicate values in this dataset
```
